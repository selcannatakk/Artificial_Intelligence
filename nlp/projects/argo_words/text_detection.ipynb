```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("data.csv")
```

```python
#--------------------------- Veri etiketleme yöntemi bu şekilde yapılacak --------------------------
```

```python

def labeling(dataset, labels):

  # tüm texler küçük harf yapılıyor
  for i in range(len(labels)-1):
      labels[i] = labels[i].lower()

  # print(labels.values().lower())
  labels = {word: index  for index, word in enumerate(labels.values())}

  for text in range(len(dataset)-1):
      dataset[text].lower()

      # liste yapılıyor
      text_dataset = dataset[text].split()

      # arayacağımız texti sözlük formatına getiriyoruz
      words_dataset = {word: index for index, word in enumerate(text_dataset)}

      # searched_word_labels -> sözlük -> 1 boyutu az
      # words_dict -> sözlük 2
      common_keys = []
      for key in labels:
          if key in words_dataset:
              common_keys.append(key)

      row = 0
      print(words_dataset)
      print(labels['say'])
      for word in words_dataset:
        row+=1
        for key in common_keys:
          if word == key:
            print(key)
            data.at[row, 'Label'] = word


  print(data.head())
  print("finish")



```

```python
dataset = data['Message']
labels = {
  0: 'say',
  1: 'ok',
  2: 'salak'}

labeling(dataset, labels)
```

```python
#--------------------------- veri ön işleme  --------------------------
```

```python

# LabelEncoder'ı oluştur
encoder = LabelEncoder()

# 'Meyve' sütununu encode et
data['Encoded_Label'] = encoder.fit_transform(data['Label'])

# Encode edilmiş DataFrame'i göster
print(data)
```

```python
# Result data (model kullanıcak)
data.to_csv('labeling_data.csv', index=False)
```

```python
# -------------------------------- Şimdi veriyi model için kullanabiliriz ---------------------
```

```python
dataset = pd.read_csv("labeling_data.csv")
```

```python
def data_split(dataset):
  x_train, x_val, y_train, y_val = train_test_split(dataset.Message, dataset.Encoded_Label, test_size=0.2, random_state=0)
  return x_train, x_val, y_train, y_val
```

```python
x_train, x_val, y_train, y_val = data_split(dataset)
```

```python
x_train.describe()
```

```python
vectorizer = CountVectorizer()
x_train_count = vectorizer.fit_transform(x_train.values)
```

```python
x_train_count.toarray()
```

```python
# ------------------------------------- Model --------------------------------
```

```python
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
#---------------------------  local test  ------------------------
```

```python
# kötü söz içermeyen text
text = [" hey wanna meet u for the game?"]
text_count = vectorizer.transform(text)
text_count.toarray()
#  label değeri sonucu verir
model.predict(text_count)
```

```python
# kötü söz içeren text
new_text = ["Ok lar... Joking wif u oni..." ]
new_text_count = vectorizer.transform(new_text)
new_text_count.toarray()
#  label değeri sonucu verir
model.predict(new_text_count)
```

```python
#---------------------------  model test  ------------------------
```

```python
# test model
x_val_count = vectorizer.transform(x_val)
# x_test_count.toarray()
model.score(x_val_count, y_val)
```
